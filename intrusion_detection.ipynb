```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE

# Load dataset
df = pd.read_csv("intrusion_detection_dataset1.csv")

# Separate features and target
X = df.drop(columns=['attack_type'])
y = df['attack_type']

# Encode target labels
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# Identify categorical and numeric columns
categorical_cols = X.select_dtypes(include='object').columns.tolist()
numeric_cols = X.select_dtypes(include='number').columns.tolist()

# Preprocessing transformer
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', 'passthrough', numeric_cols)
    ])

# Apply preprocessing to training data before SMOTE
X_train_processed = preprocessor.fit_transform(X_train)
X_test_processed = preprocessor.transform(X_test)

# Apply SMOTE to balance classes
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train_processed, y_train)

# Train Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_resampled, y_resampled)

# Predict and evaluate
y_pred = model.predict(X_test_processed)

# Results
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib
import onnx
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType

# Load dataset
df = pd.read_csv("intrusion_detection_dataset1.csv")

# Separate features and target
X = df.drop("attack_type", axis=1)
y = df["attack_type"]

# Label Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Save label encoder
joblib.dump(le, "label_encoder.joblib")

# Identify categorical and numerical columns
categorical_cols = ["protocol", "service", "flag"]
numerical_cols = [col for col in X.columns if col not in categorical_cols]

# Preprocessing
preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
    ("num", StandardScaler(), numerical_cols)
])

# Classifiers
rf = RandomForestClassifier(n_estimators=100, random_state=42)
lgb = LGBMClassifier(n_estimators=100, random_state=42)
cat = CatBoostClassifier(verbose=0, random_state=42)

ensemble = VotingClassifier(
    estimators=[("rf", rf), ("lgb", lgb), ("cat", cat)],
    voting="soft"
)

pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", ensemble)
])

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Train
pipeline.fit(X_train, y_train)

# Save pipeline
joblib.dump(pipeline, "ensemble_model.joblib")

# Evaluate
y_pred = pipeline.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Convert to ONNX
initial_type = [("float_input", FloatTensorType([None, X.shape[1]]))]
onnx_model = convert_sklearn(pipeline, initial_types=initial_type)
with open("ensemble_model.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

import skl2onnx
from skl2onnx import convert_sklearn
from skl2onnx.common.data_types import FloatTensorType, StringTensorType
from skl2onnx.common.data_types import SequenceType, DictionaryType
import onnx

# Step 1: Load data
df = pd.read_csv("intrusion_detection_dataset1.csv")

# Step 2: Define features and target
X = df.drop("attack_type", axis=1)
y = df["attack_type"]

# Step 3: Encode target
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Step 4: Identify column types
categorical_cols = ["protocol", "service", "flag"]
numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns.difference(categorical_cols).tolist()

# Step 5: Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

# Step 6: Preprocessing pipeline
preprocessor = ColumnTransformer([
    ("cat", OneHotEncoder(handle_unknown='ignore', sparse=False), categorical_cols),
    ("num", StandardScaler(), numeric_cols)
])

# Step 7: Classifiers
rf = RandomForestClassifier(n_estimators=100, random_state=42)
lgb = LGBMClassifier(n_estimators=100, random_state=42)
cat = CatBoostClassifier(verbose=0, random_state=42)

voting_clf = VotingClassifier(estimators=[
    ('rf', rf),
    ('lgb', lgb),
    ('cat', cat)
], voting='soft')

# Step 8: Pipeline
pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("classifier", voting_clf)
])

# Step 9: Fit model
pipeline.fit(X_train, y_train)

# Step 10: Evaluate
y_pred = pipeline.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy:", acc)
print("Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))

# Step 11: Export to ONNX
# Define input types: every input column must be defined!
initial_type = []

for col in categorical_cols:
    initial_type.append((col, StringTensorType([None, 1])))

for col in numeric_cols:
    initial_type.append((col, FloatTensorType([None, 1])))

# Now convert
onnx_model = convert_sklearn(pipeline, initial_types=initial_type)

# Save the model
with open("ensemble_model.onnx", "wb") as f:
    f.write(onnx_model.SerializeToString())
```
